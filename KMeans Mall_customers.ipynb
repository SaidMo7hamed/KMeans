```python
X = df.iloc[:,[3,4]].values
X.shape
```

```python
print(X)
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)    
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('Wcss')
plt.show()
```

Number of clusters=5

```python
kmeans = KMeans(n_clusters= 5, init = 'k-means++', random_state=0)
y_kmeans=kmeans.fit_predict(X)
print(y_kmeans)
```

```python
plt.scatter(X[y_kmeans==0, 0], X[y_kmeans==0, 1], s=40, c='red')
plt.scatter(X[y_kmeans==1, 0], X[y_kmeans==1, 1], s=40, c='blue')
plt.scatter(X[y_kmeans==2, 0], X[y_kmeans==2, 1], s=40, c='green')
plt.scatter(X[y_kmeans==3, 0], X[y_kmeans==3, 1], s=40, c='violet')
plt.scatter(X[y_kmeans==4, 0], X[y_kmeans==4, 1], s=40, c='yellow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan')
plt.title('Clusters')
plt.xlabel('Annual Income (k$ ')
plt.ylabel('Spending Score (1-100)')
plt.show()           
```
